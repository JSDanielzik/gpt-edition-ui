```python
import xml.etree.ElementTree as ET

# Dateipfad zur XML-Datei
xml_datei_pfad = '00000002.xml'

# Namespace
namespace = {'page': 'http://schema.primaresearch.org/PAGE/gts/pagecontent/2019-07-15'}

# Erstelle eine leere Liste
ergebnis_liste = []

# Parse die XML-Datei
tree = ET.parse(xml_datei_pfad)
root = tree.getroot()

# Iteriere durch die TextLine-Elemente
for text_line in root.findall('.//page:TextLine', namespaces=namespace):
    # Extrahiere den Text
    text_equiv = text_line.find('./page:TextEquiv/page:Unicode', namespaces=namespace)
    if text_equiv is not None and text_equiv.text:
        text = text_equiv.text.strip()

        # Extrahiere die Koordinaten
        coords = text_line.find('./page:Coords', namespaces=namespace)
        if coords is not None and 'points' in coords.attrib:
            coords_raw = coords.attrib['points'].split(" ")
            #print(coords_raw)
            coords_data = []
            for c in coords_raw:
                #print(c)
                coords_data.append({'x': c.split(',')[0], 'y': c.split(',')[1]})

            # Füge ein Dictionary zur Liste hinzu
            ergebnis_liste.append({'text': text, 'coords': coords_data})

# Ergebnis ausgeben
for eintrag in ergebnis_liste:
    print(eintrag)

obj = {'page-scan': 'https://www.digitale-bibliothek-mv.de/viewer/api/v1/records/PPN807078646_1904/files/images/00000002.tif/full/!2048,2048/0/default.jpg?ignoreWatermark=false&watermarkText=Universit%C3%A4t+Greifswald',
       'segments': ergebnis_liste}
with open("api-mock-with-coords.json", 'wb') as file:
    file.write(json.dumps(obj, ensure_ascii=False).encode('utf-8'))
```
